# Clean and Merge Donation and Project Opendata

```python
import pandas as pd
```

### Clean the donations file

```python
#read donations file with donor info
donations = pd.read_csv('opendata/opendata_donations000', sep=',', dtype={'donor_zip': object}, 
                                        names=['_donationid', '_projectid', '_donor_acctid', '_cartid', 'donor_city', 'donor_state', 'donor_zip', 'is_teacher_acct', 'donation_timestamp', 'donation_to_project', 'donation_optional_support', 'donation_total', 'donation_included_optional_support', 'payment_method', 'payment_included_acct_credit', 'payment_included_campaign_gift_card', 'payment_included_web_purchased_gift_card', 'payment_was_promo_matched', 'is_teacher_referred', 'giving_page_id', 'giving_page_type', 'for_honoree', 'thank_you_packet_mailed'])
```

```python
donations.head()
```

```python
donations.size
```

```python
donations_cleaned=donations[['donation_timestamp', '_donationid', '_projectid', '_donor_acctid', 'donor_city', 'donor_state', 'donor_zip', 'donation_total']]
```

```python
#drop any donor without zip info
donations_cleaned=donations_cleaned.dropna(subset=['donor_zip'])
```

### Clean the Project file

```python
#projects file with project info
projects = pd.read_csv('openData/opendata_projects000', sep=',', dtype={'school_zip': object},
                       names=['_projectid', '_teacher_acctid', '_schoolid', 'school_ncesid', 'school_latitude', 'school_longitude', 'school_city', 'school_state', 'school_zip', 'school_metro', 'school_district', 'school_county', 'school_charter', 'school_magnet', 'school_year_round', 'school_nlns', 'school_kipp', 'school_charter_ready_promise', 'teacher_prefix', 'teacher_teach_for_america', 
                                                                          'teacher_ny_teaching_fellow', 'primary_focus_subject', 'primary_focus_area' ,'secondary_focus_subject', 'secondary_focus_area', 'resource_type', 'poverty_level', 'grade_level', 'vendor_shipping_charges', 'sales_tax', 'payment_processing_charges', 'fulfillment_labor_materials', 'total_price_excluding_optional_support', 'total_price_including_optional_support', 'students_reached', 'total_donations', 'num_donors', 'eligible_double_your_impact_match', 'eligible_almost_home_match', 'funding_status', 'date_posted', 'date_completed', 'date_thank_you_packet_mailed', 'date_expiration'])
```

```python
projects.head()
```

```python
#eliminte project columns
projects_cleaned=projects[["_projectid", "_schoolid", "school_latitude", 
                           "school_longitude", "school_zip", "school_city", "total_donations", "num_donors"]]

```

```python
#drop any school without zip info
projects_cleaned=projects_cleaned.dropna(subset=['school_zip'])
```

```python
projects_cleaned.head()
```

### Merge donation with projects

```python
donor_charity_combined=donations_cleaned.merge(projects_cleaned, left_on="_projectid", right_on="_projectid", how="left")
```

### Label lastday and givingtuesday and trim down from 2007-2017

```python
donor_charity_combined['date_stamp']=donor_charity_combined.donation_timestamp.apply(lambda x: x.split(' ')[0])
```

```python
donor_charity_combined['year']=donor_charity_combined.date_stamp.apply(lambda x: int(x.split('-')[0]))
donor_charity_combined['month']=donor_charity_combined.date_stamp.apply(lambda x: int(x.split('-')[1]))
donor_charity_combined['day']=donor_charity_combined.date_stamp.apply(lambda x: int(x.split('-')[2]))
```

```python
donor_charity_combined=donor_charity_combined[donor_charity_combined.year>2006]
```

```python
"""
Define a lambda function f that return 1 if the date is within the set of giving tuesday else 0.
"""

def f(x):
    return 1 if x in ['2006-11-28', '2007-11-27', '2008-12-02', '2009-12-11', 
                         '2010-11-30', '2011-11-29', '2012-11-27', '2013-12-03', 
                         '2014-12-02', '2015-12-01', '2016-11-29'] else 0

f('2012-11-27')
```

```python
"""
Define a lastday function that return 1 if the date is the last day of the zero.
"""

def lastday(x):
    return 1 if x in ['2006-12-31', '2007-12-31', '2008-12-31', '2009-12-31', 
                         '2010-12-31', '2011-12-31', '2012-12-31', '2013-12-31', 
                         '2014-12-31', '2015-12-31', '2016-12-31'] else 0

lastday('2012-12-31')
```

```python
donor_charity_combined['GivingTuesday']=donor_charity_combined.date_stamp.apply(f)
donor_charity_combined['LastDay']=donor_charity_combined.date_stamp.apply(lastday)
```

```python
donor_charity_combined[donor_charity_combined.date_stamp=='2010-11-30']
```

### Write to file 

```python
donor_charity_combined.to_csv("donations_projects_merged_with_location_amount_updated_Open.csv")
```

## Calculate Daily Contribution sum and count

```python
#groupby date
small_data=donor_charity_combined.groupby('date_stamp')
```

```python
#get only the date and donation amount
df_small=donor_charity_combined[['date_stamp', 'donation_total', 'year', 'month', 'day']]
```

```python
df_small['donation_total']=df_small.donation_total.astype('float')
```

#### Get the donation sum and the count per date dataframes and merge them

```python
daily_donation=df_small.groupby(["date_stamp"]).sum().reset_index()
```

```python
daily_donation_count=df_small.groupby(["date_stamp"]).count().reset_index()
daily_donation_count['donation_count']=daily_donation_count['donation_total']
daily_donation_count=daily_donation_count[['date_stamp', 'donation_count']]
daily_donation_count.head()
```

```python
daily_donation=daily_donation[['date_stamp', 'donation_total']].merge(daily_donation_count, left_on="date_stamp", right_on="date_stamp", how="left")
```

#### Create the GivingTuesday and LastDay Columns to the new dataframe and verify

```python
daily_donation['GivingTuesday']=daily_donation.date_stamp.apply(f)
daily_donation[daily_donation.date_stamp=='2012-11-27']
```

```python
daily_donation['LastDay']=daily_donation.date_stamp.apply(lastday)
daily_donation.head()
```

```python
daily_donation[daily_donation.GivingTuesday==1]
```

### Write to file

```python
daily_donation.to_csv('daily_donation_Open.csv')
```
